```python
import re

def remove_tags_hashtags(text):
    """Removes account tags (@username) and hashtags (#hashtag) from text."""
    if isinstance(text, str):  # Check if the input is a string
        text = re.sub(r'@\w+', '', text)
        text = re.sub(r'#\w+', '', text)
    return text  # Return the original value if it's not a string

# Apply the modified function to the 'text' column
df['text'] = df['text'].apply(remove_tags_hashtags)
```

```python
def remove_emojis(text):
    """Removes emojis from the text."""
    # This regular expression matches all emojis and other symbols

    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F700-\U0001F77F"  # alchemical symbols
        "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        "\U0001FA00-\U0001FA6F"  # Chess Symbols
        "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        "\U00002702-\U000027B0"  # Dingbats
        "\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE
    )
    if isinstance(text,str):
      text = emoji_pattern.sub(r'', text)
    return text

# Example usage with a DataFrame column
df['text'] = df['text'].apply(remove_emojis)
```

```python
import pandas as pd
import re
import gensim
from gensim import corpora
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
df = df[df['text'].apply(lambda x: isinstance(x, str))]
```

```python
df.shape
```

```python
def preprocess_text(text):
    # Lowercase the text
    text = text.lower()
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove @mentions and hashtags
    text = re.sub(r'@\w+|#\w+', '', text)
    # Remove special characters, numbers, and punctuation
    text = re.sub(r'\W|[\d_]', ' ', text)
    # Tokenize and remove stopwords
    words = word_tokenize(text)
    words = [word for word in words if word not in stopwords.words('english') and len(word) > 2]
    return words

# Apply preprocessing to the tweet text
df['processed_text'] = df['text'].apply(preprocess_text)
```

```python
dictionary = corpora.Dictionary(df['processed_text'])

# Filter out rare and common words
dictionary.filter_extremes(no_below=5, no_above=0.5)

# Convert documents to Bag-of-Words format
corpus = [dictionary.doc2bow(text) for text in df['processed_text']]
```

```python
# Set the number of topics
num_topics = 5  # Change as needed for your analysis

# Train the LDA model
lda_model = gensim.models.LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, random_state=42, passes=10, alpha='auto')

# Print topics with words and their weights
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx + 1}: {topic}")
```
